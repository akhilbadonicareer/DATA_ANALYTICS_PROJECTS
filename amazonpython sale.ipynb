```python
df_numer=df
for col in df_numer.columns:
    if(df_numer[col].dtype=='object'):
        df_numer[col]=df_numer[col].astype('category')
        df_numer[col]=df_numer[col].cat.codes
```

```python
df_numer.corr()
```

```python
corr_mat1= df_numer.corr()
corr_pairs=corr_mat1.unstack()
corr_pairs
sp=corr_pairs.sort_values()
high_corr= sp[sp>0.5]
high_corr
```

## from this we found out there is a huge correlation between Courier Status and Qty
